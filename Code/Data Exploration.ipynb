### Step 1: Pre requisite functions

#### This function load the data from the given path and finename

```python
#Function to load the data
def loadData(path,filename,sepr):
    try:
             files = os.listdir(path)
             for f in files:
                 if f == filename:
                     data = pd.read_csv(os.path.join(path,f),sep=sepr)
                     return data
            
    except Exception as ex:
           print ("-----------------------------------------------------------------------")
           template = "An exception of type {0} occurred. Arguments:\n{1!r}"
           message = template.format(type(ex).__name__, ex.args)
           print (message)
```

#### This function derives the shape of the dataset and returns the feature and target columns

```python
#Function to explore the data
def exploreData(data):
    try:
           #Total number of records                                  
           rows = data.shape[0]
           cols = data.shape[1]    
           
           #separate features and target
           drop_col = ['target']
           features = data.drop(drop_col, axis = 1)
           target = data[drop_col]
          
           # Print the results
           print ("-----------------------------------------------------------------------")
           print ("Total number of records: {}".format(rows))
           print ("Total number of features: {}".format(cols))
           print ("-----------------------------------------------------------------------")
           
           return features,target
           
    except Exception as ex:
           print ("-----------------------------------------------------------------------")
           template = "An exception of type {0} occurred. Arguments:\n{1!r}"
           message = template.format(type(ex).__name__, ex.args)
           print (message)
```

#### This function computes the percentage of missing values per each column in the given dataset

```python
def missingValues(data):
    try:
           # Total missing values
           mis_val = data.isnull().sum()
         
           # Percentage of missing values
           mis_val_percent = 100 * mis_val / len(data)
           
           # Make a table with the results
           mis_val_table = pd.concat([mis_val, mis_val_percent], axis=1)
           
           # Rename the columns
           mis_val_table_ren_columns = mis_val_table.rename(
           columns = {0 : 'Missing Values', 1 : '% of Total Values'})
           mis_val_table_ren_columns.head(4 )
           # Sort the table by percentage of missing descending
           misVal = mis_val_table_ren_columns[
            mis_val_table_ren_columns.iloc[:,1] != 0].sort_values(
                   '% of Total Values', ascending=False).round(1)
                     
           return misVal, mis_val_table_ren_columns

    except Exception as ex:
           print ("-----------------------------------------------------------------------")
           template = "An exception of type {0} occurred. Arguments:\n{1!r}"
           message = template.format(type(ex).__name__, ex.args)
           print (message)
```

#### This function plots the counts of a value per  target feature classification. Feature should be categorical.

```python
def catCount(feature,target,data): 
    try:
        d_f = data.loc[data[target] == False]
        d_t = data.loc[data[target] == True]
        
        d_f[feature].fillna(value='missing',inplace=True)
        d_t[feature].fillna(value='missing',inplace=True)
         
        f, axes = plt.subplots(1, 2, figsize=(8, 8), sharex=True)
        sns.countplot(x=feature, data=d_f,ax=axes[0])
        axes[0].set_title('target=False')
        sns.countplot(x=feature, data=d_t,ax=axes[1])
        axes[1].set_title('target=True')    
        for ax in axes:
            ax.set_xticklabels(ax.get_xticklabels(), rotation=65, horizontalalignment='right')                
        plt.show()
    except Exception as ex:
        print ("-----------------------------------------------------------------------")
        template = "An exception of type {0} occurred. Arguments:\n{1!r}"
        message = template.format(type(ex).__name__, ex.args)
        print (message) 
```

#### This column derives the density probability function for continous features.

```python
def numCount(feature,target,data):
    try:
        d_f = data.loc[data[target] == False]
        d_t = data.loc[data[target] == True]
         
        plt.figure(figsize = (12, 6))
        sns.kdeplot(d_f[feature], label='target=false')
        sns.kdeplot(d_t[feature], label='target=true')
        plt.title(feature)
        plt.legend();
    except Exception as ex:
        print ("-----------------------------------------------------------------------")
        template = "An exception of type {0} occurred. Arguments:\n{1!r}"
        message = template.format(type(ex).__name__, ex.args)
        print (message)
```

### Step 3: Combine all the datasets

```python
#Add columns to data frames to make columns common across
#Extract list of columns from dataset1,2,3
col1 = list(data1.columns)
col2 = list(data2.columns)
col3 = list(d_f.columns)

#Concatenate all the columns
tot_col = col1 + col2 + col3
tot_col = list(dict.fromkeys(tot_col))

#Compile the list of extra columns to be added for each dataset
c1 = list(set(tot_col) - set(col1))
c2 = list(set(tot_col) - set(col2))
c3 = list(set(tot_col) - set(col3))

#Add the derived list of columns to corresponding dataset
for i in c1:
    data1[i] = 0

for i in c2:
    data2[i] = 0

for i in c3:
    d_f[i] = 0

#Concatenate three datasets in to one
data = pd.concat([data1,data2,d_f], axis=0)

print ("------------------------Combined Heart dataset-------------------------------")
misVal, mis_val_table_ren_columns = missingValues(data)
# Print some summary information
print ("Columns that have missing values:" + str(misVal.shape[0]))
print ("-----------------------------------------------------------------------")
print(mis_val_table_ren_columns.head(20))
```

Analysis:
1. We have only one column 'age' which is a common column among all the three datasets 
2. But above column is a continous feature and thus cannot be used a key to combine the three datasets.
3. As we do not have a key, we can add default values to the missing columns and create a combined datasets with common set of columns.
